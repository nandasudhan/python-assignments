Scipy:
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Given data
months = np.arange(12)
max_temp = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
min_temp = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
```

```python
def periodic_func(x, a, b, c, d):
    return a * np.sin(b * x + c) + d
```

```python
popt_max, pcov_max = curve_fit(periodic_func, months, max_temp)
```

```python
popt_min, pcov_min = curve_fit(periodic_func, months, min_temp)
```

```python
# Generate data points for plotting the fit
months_fit = np.linspace(0, 11, 100)
```

```python
max_temp_fit = periodic_func(months_fit, *popt_max)
```

```python
min_temp_fit = periodic_func(months_fit, *popt_min)
```

```python
# Plot the original data and the fit
plt.figure(figsize=(10, 6))
plt.plot(months, max_temp, 'ro', label='Max Temp Data')
plt.plot(months, min_temp, 'bo', label='Min Temp Data')
plt.plot(months_fit, max_temp_fit, 'r-', label='Max Temp Fit')
plt.plot(months_fit, min_temp_fit, 'b-', label='Min Temp Fit')
plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.title('Fitting to Periodic Function')
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
```

```python
titanic
```

```python
#Create a pie chart presenting the male/female proportion
```

```python
# 1. Create a pie chart presenting the male/female proportion
gender_proportion = titanic['sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_proportion, labels=gender_proportion.index, autopct='%1.1f%%', startangle=140)
plt.title('Male/Female Proportion on Titanic')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
plt.figure(figsize=(10, 6))
plt.scatter(titanic[titanic['sex'] == 'male']['age'], titanic[titanic['sex'] == 'male']['fare'], color='blue', label='Male', alpha=0.5)
plt.scatter(titanic[titanic['sex'] == 'female']['age'], titanic[titanic['sex'] == 'female']['fare'], color='red', label='Female', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare Paid')
plt.title('Scatterplot of Fare Paid vs Age (Colored by Gender)')
plt.legend()
plt.grid(True)
plt.show()
```
